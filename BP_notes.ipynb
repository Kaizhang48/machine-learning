### 为什么使用Backpropagation
本质上，Backpropagation是chain rule的扩展，主要用于求出神经网络中目标函数对权重的偏导，从而可以求出最优化结果的一种方法。

- cost function
$$Cost = \frac{1}{2 \, N} \, \sum^{N}_{i=1} (h(x^{(i)})-y^{i})^2$$

- update params
$$\Theta = \Theta - \eta \, \frac{\partial{Cost}}{\partial{\Theta}}$$

### 简单的Backpropagation 
下图是$e=(a+b)(b+1)$的关系图，包含了加法门和乘法门。



利用链式法则我们知道：
$$
\frac{\partial e}{\partial a}=\frac{\partial e}{\partial c}\cdot \frac{\partial c}{\partial a} \\
\frac{\partial e}{\partial b}=\frac{\partial e}{\partial c}\cdot \frac{\partial c}{\partial b}+\frac{\partial e}{\partial d}\cdot \frac{\partial d}{\partial b}
$$

我们可以自上而下算出偏导，这也是Backpropagation的简单的应用。
